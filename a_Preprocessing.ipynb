# Daten verarbeiten
## Variablen festlegen

```python
#Remove labels
features = data_dummies.drop(['mehrals50K', 'wenigerals50K'], axis=1)
labels = data_dummies.mehrals50K
features_train, features_test, labels_train, labels_test = train_test_split(features, labels, test_size = 0.25, random_state=5)
print(features_train.shape)
print(labels_train.shape)
print(features_test.shape)
print(labels_test.shape)
```

## Baseline erstellen

```python
feature_list = list(features.columns)
#baseline_preds = features_test[:, feature_list.index('mehrals50K')]
```

# Random Forest

```python
clf = RandomForestClassifier()
clf.fit(features_train, labels_train)
```

```python
predictions = clf.predict(features_test)
predictions
```

## Genauigkeit überprüfen 

```python
print ("Test Accuracy  :: ", accuracy_score(labels_test, predictions))
print ("Confusion matrix ", confusion_matrix(labels_test, predictions))
```

# Tree

```python
clf_gini = DecisionTreeClassifier(criterion = "gini", random_state = 100,
                               max_depth=3, min_samples_leaf=5)
clf_gini.fit(features_train, labels_train)
```

```python
y_pred = clf_gini.predict(features_test)
y_pred
```

## Genauigkeit überprüfen

```python
print ("Test Accuracy  :: ", accuracy_score(labels_test, y_pred))
print ("Confusion matrix ", confusion_matrix(labels_test, y_pred))
```

# Gradient Boosting Classifier

```python
#Create a svm Classifier
gb = GradientBoostingClassifier(n_estimators=20, learning_rate = 1, max_features=2, max_depth = 2, random_state = 0)
gb.fit(features_train, labels_train)

#Predict the response for test dataset
z_pred = gb.predict(features_test)
z_pred
```

## Genauigkeit überprüfen

```python
print ("Test Accuracy  :: ", accuracy_score(labels_test, z_pred))
print ("Confusion matrix ", confusion_matrix(labels_test, z_pred))
matrix = confusion_matrix(labels_test, z_pred)



tpr, fpr, thresholds = metrics.roc_curve(labels_test, z_pred, pos_label=0)
print((fpr))
print()
print(tpr)
# Print ROC curve
plt.plot(tpr,fpr)
plt.show() 

# Print AUC
auc = np.trapz(fpr,tpr)
print('AUC:', auc)

plt.show()
```
