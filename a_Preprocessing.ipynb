# One-Hot-Kodierung

```python
print(data.gender.value_counts())
```

Kategorien aus Spalten herausziehen -> zu neuen Zeilen mit binären Werten machen

```python
print("Original features:\n", list(data.columns), "\n")
data_dummies = pd.get_dummies(data)
print("Features after get_dummies:\n", list(data_dummies.columns))
```

```python
data_dummies.head()
```

```python
data_dummies.rename(columns={'income_>50K': 'mehrals50K'}, inplace=True)
data_dummies.keys()
```

```python
data_dummies.to_csv('Adult_preprocessed.csv', encoding='utf-8', index=False)
```
